# Preprocessing

```python
label = [' yummy','convenient','spicy','fattening','greasy','fast','cheap','tasty','expensive','healthy','disgusting','VisitFrequency','Gender']
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in label:
    data[i] = le.fit_transform(data[i])
data
```

```python
data.replace('[a-zA-z+!]','',regex=True,inplace=True)
data.head()
```

```python
import numpy as np
data['Like'] = np.int64(data['Like'])
data['Like'].dtype
```

# PCA

```python
# Apply PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=None)
pca.fit(data)

# Get the eigenvalues
print("Eigenvalues:")
print(pca.explained_variance_)
print()

# Get explained variances
print("Variances (Percentage):")
print(pca.explained_variance_ratio_ * 100)
print()

# Make the scree plot
plt.plot(np.cumsum(pca.explained_variance_ratio_ * 100))
plt.xlabel("Number of components (Dimensions)")
plt.ylabel("Explained variance (%)")
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
prinComp = pca.fit_transform(data)
```

```python
pca_df = pd.DataFrame(data = prinComp, columns = ['principal component 1','principal component 2'])
```

```python
pca_df
```

```python
print('Explained variation per principal component: {}'.format(pca.explained_variance_ratio_))
```

```python
sns.heatmap(pca_df.corr(), annot=True)
```

# Clustering

```python
from sklearn.cluster import KMeans
cluster_sumofsquare = [] # Will save the values of all clusters sum of squares, which means how far the observed values is from centroid

for k in range(1,11):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(pca_df)
    cluster_sumofsquare.append(kmeans.inertia_)

plt.figure(figsize=(12,6))    
plt.plot(cluster_sumofsquare)
plt.xlabel("K Value")
plt.ylabel("WCSS")
plt.show()
```

```python
model = (KMeans(n_clusters = 5) )

model.fit(pca_df)
model
```

```python
labels2 = model.labels_
centroids2 = model.cluster_centers_

print(labels2)  #Labels of each point, means to which cluster they belong
print(centroids2) #centroids of each cluster in both columns
```

```python
#Clusters in 3d projection

pca_df['label'] = labels2 

fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(pca_df["principal component 1"][pca_df.label == 0], pca_df["principal component 2"][pca_df.label == 0], c='blue', s=60)
ax.scatter(pca_df["principal component 1"][pca_df.label == 1], pca_df["principal component 2"][pca_df.label == 1], c='red', s=60)
ax.scatter(pca_df["principal component 1"][pca_df.label == 2], pca_df["principal component 2"][pca_df.label == 2], c='green', s=60)
ax.scatter(pca_df["principal component 1"][pca_df.label == 3], pca_df["principal component 2"][pca_df.label == 3], c='orange', s=60)
ax.scatter(pca_df["principal component 1"][pca_df.label == 4], pca_df["principal component 2"][pca_df.label == 4], c='purple', s=60)
ax.view_init(30, 210)
plt.xlabel("principal component 1")
plt.ylabel("principal component 2")

plt.show()
```

```python
km_5 = KMeans(n_clusters=5, init='k-means++', random_state=0)
km_5.fit(pca_df)
centroids = km_5.cluster_centers_
pca_df['Labels'] = km_5.labels_


plt.figure(figsize=(10, 6))

sns.scatterplot(pca_df['principal component 1'], pca_df['principal component 2'], hue=pca_df['Labels'], 
                palette=sns.color_palette('hls', 5))

plt.scatter(centroids[:,0], centroids[:,1], c='red',s=200)

plt.title('K-Means Clustering',fontsize=18)
plt.show()
```
